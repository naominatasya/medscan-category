```python
# Pisahkan data berdasarkan frekuensi kategori
kategori_counts = filtered_df['Kategori Produk Baru Final'].value_counts()

# Kategori dengan jumlah data >= 100
kategori_banyak = kategori_counts[kategori_counts >= 100].index.tolist()
df_banyak = filtered_df[filtered_df['Kategori Produk Baru Final'].isin(kategori_banyak)]

# Kategori dengan jumlah data < 100
kategori_dikit = kategori_counts[kategori_counts < 100].index.tolist()
df_dikit = filtered_df[filtered_df['Kategori Produk Baru Final'].isin(kategori_dikit)]

# Split hanya untuk kategori dengan jumlah data >= 100
df_banyak_70, df_banyak_30 = train_test_split(df_banyak, test_size=0.3, random_state=42, stratify=df_banyak['Kategori Produk Baru Final'])

# Simpan hasil
df_banyak_70.to_excel(r"C:\Users\Naomi Natasya\Downloads\filtered_kategori_produk_banyak_70.xlsx", index=False)
df_banyak_30.to_excel(r"C:\Users\Naomi Natasya\Downloads\filtered_kategori_produk_banyak_30.xlsx", index=False)
df_dikit.to_excel(r"C:\Users\Naomi Natasya\Downloads\filtered_kategori_dikit.xlsx", index=False)
```
